#### Ejercicio 1

(i) Define una caminata aleatoria en 2D, con coordenadas $x$ y $y$, que vive en $\mathbb{Z}^2$.

(ii) Escribe una función que hace un paso del caminante. Puede moverse sólo arriba, abajo, a la izquierda o a la derecha en cada paso, con probabilidades iguales.

(iii) Escribe una simulación de $N$ caminantes en 2D, siguiendo las mismas ideas que en el notebook anterior.

```julia
function paso()
    r=rand(Bool)
    s=Int(r)
        if s==0
            D=-1
        elseif s==1
            D=1
        end #para mayyor comodidad lo dividimos en 2 funciones
    return D
end
```

```julia
function caminatas(T)
    posicionesx=[]
    posicionesy=[]
    x=0
    y=0
    p=0
    for i in 1:T
      push!(posicionesx,x)
      push!(posicionesy,y)
      d=rand(Bool)
        if d==false
      p=paso()
            x=x+p
        elseif  d==true
      p=paso()
            y=y+p
        end
    end
    return posicionesx, posicionesy 
end
```

```julia
posicionesx, posicionesy=caminatas(1000)
```

```julia
using Plots
```

```julia
plot(posicionesx, posicionesy)
```

#### Ejercicio 3

(i) Calcula el promedio y la varianza como función del tiempo y grafícalos.

```julia
#promedio 
function promediox(T)
suma=0
v,u=caminatas(T)
l=length(v)
    for i in 1:T
        suma=suma+v[i]+u[i]
    end
    p1=suma/l
    return p1
end  
```

```julia
promediox(100)
```

```julia
function promediosx(T)
    return [promediox(T) for i in 1:T]
end
```

```julia
function pp2dx(N,T)

pc=plot(leg=false)
    for i in 1:N
    a=promediosx(T)
       pc= plot!(a)
    end
plot(pc,xlabel="Tiempo",ylabel="Promedio")
end
```

```julia
pp2dx(3,100)
```

```julia
#ii varianza
function varianza(T)
    u,v=caminatas(T)
    n = length(v)
    prom = promediox(T)
    
    suma = 0
    
    for i in 1:n
        h[i]=u[i]+v[i]
        t = (h[i]-prom)^2
        suma = suma + t
    end
        
    return suma/n
end
```

```julia
varianza(10)
```

```julia
function varianzas(T)
    return [varianza(T) for i in 1:T]
end
```

```julia
varianzas(10)
```

```julia
function pv2d(N,T)

pc=plot(leg=false)
    for i in 1:N
    a=varianzas(T)
       pc= plot!(a)
    end
plot(pc,xlabel="Tiempo",ylabel="Promedio")
end
```

```julia
pv2d(3,10)
```

```julia
function caminatas3d(T)
    posicionesx=[]
    posicionesy=[]
    posicionesz=[]
    x=0
    y=0
    z=0
    p=0
    for i in 1:T
      push!(posicionesx,x)
      push!(posicionesy,y)
      push!(posicionesz,z)
      d=rand()
      p=paso()
        if 0<=d<=1/3
            x=x+p
        elseif  1/3<d<=2/3
            y=y+p
        elseif 2/3<d<=1
           z=z+p
        end
    end
    return posicionesx, posicionesy, posicionesz 
end
```

```julia
px,py,pz=caminatas3d(1000)
```

```julia
plot(px,py,pz)
```

```julia
#i
function uniforme() 
    b=rand([-1,1])
    a=b*rand()
    return a
end
```

```julia
function vect(n)
    v=[]
    for i in 1:n
        push!(v,uniforme())
    end
    return v
end
        
```
