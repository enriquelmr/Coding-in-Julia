```julia
a=[]
b=[]
r=round(360*rand()) #angulo
s=Int(r)#convierte el angulo en un entero
l=3 #longitud del paso
x0=0
y0=0
x=x0+l*cosd(s)
y=y0+l*sind(s)
push!(a,x)
push!(b,y)
```

```julia
function caminante(n)    
    a=[0.]
    b=[0.]
    c=[]
    x=0
    y=0
    l=2 #longitud del paso
    @gif for i in range(1,n)
        r=round(360*rand()) #angulo
        s=Int(r)#convierte el angulo en un entero
        x=x+l*cosd(s)
        y=y+l*sind(s)
        push!(a,x)
        push!(b,y)
    end
end
```

```julia
plot(a,b,m='o')
```
